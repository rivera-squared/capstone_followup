## Mean Comparison Tests

### Comparing the META-PR (Math, Spanish, Science, and English) difference in means of schools receiving displaced students and those schools not receiving displaced students. All primary schools for Academic year 2020-21

```python
%%stata -d df
ttest espanol_meta_21, by(escuela_receptora)
ttest ciencias_meta_21, by(escuela_receptora)
ttest ingles_meta_21, by(escuela_receptora)
ttest mate_meta_21, by(escuela_receptora)
ttest matricula_21, by(escuela_receptora)
```

### Comparing the META-PR (Math, Spanish, Science, and English) difference in means for Academic year 2018-19 and 2020-21. All primary schools

```python
%%stata -d df
ttest mate_meta_21 == mate_meta_19 
ttest espanol_meta_21 == espanol_meta_19 
ttest ingles_meta_21 == ingles_meta_19 
ttest ciencias_meta_21 == ciencias_meta_19 
ttest matricula_21 == matricula_2019
```

### Comparing the META-PR (Math, Spanish, Science, and English) difference in means for Academic year 2018-19 and 2020-21. All primary schools receiving displaced students

```python
%%stata -d df
ttest mate_meta_21 == mate_meta_19 if escuela_receptora == 1
ttest espanol_meta_21 == espanol_meta_19 if escuela_receptora == 1
ttest ingles_meta_21 == ingles_meta_19 if escuela_receptora == 1
ttest ciencias_meta_21 == ciencias_meta_19 if escuela_receptora == 1
ttest matricula_21 == matricula_2019 if escuela_receptora == 1
```
